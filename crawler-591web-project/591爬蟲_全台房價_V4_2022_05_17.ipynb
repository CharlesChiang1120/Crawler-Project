```python
def name_func(func):
    def wrapper(*args, **kwargs):
      print("running: ", func.__name__+"()")
      func(*args, **kwargs)
    return wrapper

def time_func(func):
    def wrapper(*args, **kwargs):
      start = time.time()
      func(*args, **kwargs)
      end = time.time()
      print("elapsed: ", round(end-start, 2), "seconds")
    return wrapper
```

```python
# values are duplicate
# inherit dictionary to make dictionary value can be list type
class Dictlist(dict):
    def __setitem__(self, key, value):
        try:
            self[key]
        except KeyError:
            super(Dictlist, self).__setitem__(key, [])
        self[key].append(value)

# class
class Crawler:
  # property
  def __init__(self, url, pagenum):
    self.url = url
    self.pagenum = pagenum
    self.headers = generate_random_header()

    # enqueue the whole url 
    self.crawler_url_queue = queue.Queue()

    # put crawler result into list to convert to dataframe 
    self.crawler_data_list = []
  
  @property
  def pagenum(self):
    return self._pagenum
  
  @pagenum.setter
  def pagenum(self, pagenum):
    if pagenum > 120:
      self._pagenum = 120
    elif pagenum < 0:
      self._pagenum = 1
    else:
      self._pagenum = pagenum

  def __repr__(self):
    return "Methods : <enqueue_url(), get_head_queue(), multithread_work()>"


  # put section id into queue
  @time_func
  @name_func
  def enqueue_url(self):
    section_dict = self._get_section_rid_list()
    for value in section_dict:
      for n in range(0, self.pagenum):
        self.crawler_url_queue.put(self.url+str(value)+"&page="+str(n))
        
  # as i don't konw how many region ids in 591
  # test all possible answer (30 records)
  # if the result is blank in a row than break the process
  def _get_section_rid_list(self):
    section_dict = Dictlist()
    count = 0
    for i in range(1, 30):
      request_url = self.url+str(i)
      response = requests.get(request_url, headers=self.headers)
      response = response.json()
      if len(response["data"]["items"]) != 0:
        section_dict[response["data"]["items"][0]["region"]] = i
        count = 0
      else:
        count += 1
        pass
      if count >= 2:
        break
    flatten_list = list(chain.from_iterable(section_dict.values()))
    section_dict = flatten_list
    return section_dict 

  # print queue elements
  def get_head_queue(self):
    print("Queue:\n")
    if self.crawler_url_queue.qsize() != 0:
      for i in range(5):
        print(self.crawler_url_queue.queue[i])
    else:
      print("Queue is empty!")
      return

  # start crawlering
  # append records into list
  def worker(self):
    try:
      while self.crawler_url_queue.qsize() > 0: 
          request_url = self.crawler_url_queue.get() 
          headers = generate_random_header()
          response = requests.get(request_url, headers=self.headers)
          response = response.json()
          for datanum in range(0, len(response["data"]["items"])):
            if len(response["data"]["items"]) != 0:
              self.crawler_data_list.append([response["data"]["items"][datanum]["address"], response["data"]["items"][datanum]["area"], response["data"]["items"][datanum]["build_name"],\
                            str('https://newhouse.591.com.tw/home/housing/info?hid='+str(response["data"]["items"][datanum]["hid"])),\
                            response["data"]["items"][datanum]["price"], response["data"]["items"][datanum]["price_unit"], response["data"]["items"][datanum]["purpose_str"],\
                            response["data"]["items"][datanum]["region"], response["data"]["items"][datanum]["regionid"], response["data"]["items"][datanum]["room"],\
                            response["data"]["items"][datanum]["section"], response["data"]["items"][datanum]["sectionid"], response["data"]["items"][datanum]["status"],\
                            response["data"]["items"][datanum]["tag"], response["data"]["items"][datanum]["updatetime"]])
            else: 
              break
    except Exception as e:
      print(e)

  @time_func
  @name_func
  def multithread_work(self):
    try:
      with ThreadPoolExecutor(max_workers=10) as executor:
        for i in range(5000):
          executor.submit(self.worker) 
    except Exception as e:
      print(e)
```
